```python
#Part1
A = np.array([[3, 4],[7, 8]])

B = np.array([[5, 3],[2, 1]])
```

```python
#1
A_inv = np.linalg.inv(A)
result = np.dot(A, A_inv)
result_rounded = np.round(result, decimals=10)
print(result_rounded)
```

```python
#2
AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB =\n", AB)
print("BA =\n", BA)
```

####1. Element-wise Addition:
• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.

```python
# Create two lists of size 1,000,000
list1 = list(range(1000000))
list2 = list(range(1000000))

# Start time
start_time = time.time()

# Element-wise addition using list comprehension
result_list = [a + b for a, b in zip(list1, list2)]

# End time
end_time = time.time()

print("Time taken using Python lists:", end_time - start_time, "seconds")
```

```python
# Create two NumPy arrays of size 1,000,000
arr1 = np.arange(1000000)
arr2 = np.arange(1000000)

# Start time
start_time = time.time()

# Element-wise addition using NumPy
result_array = arr1 + arr2

# End time
end_time = time.time()

print("Time taken using NumPy arrays:", end_time - start_time, "seconds")
```

####3. Dot Product
• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.

```python
# Create two lists of size 1,000,000
list1 = list(range(1, 1000001))
list2 = list(range(1, 1000001))

# Start timer
start_time = time.time()

# Dot product using a loop or sum of element-wise multiplication
dot_product = sum(a * b for a, b in zip(list1, list2))

# End timer
end_time = time.time()

print("Dot product using Python lists:", dot_product)
print("Time taken using Python lists:", end_time - start_time, "seconds")
```

```python
# Create two NumPy arrays
arr1 = np.arange(1, 1000001)
arr2 = np.arange(1, 1000001)

# Start timer
start_time = time.time()

# Dot product using NumPy
dot_product_np = np.dot(arr1, arr2)

# End timer
end_time = time.time()

print("Dot product using NumPy arrays:", dot_product_np)
print("Time taken using NumPy arrays:", end_time - start_time, "seconds")
```

####4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
import random
# Create two 1000x1000 matrices using Python lists
size = 1000
matrix1 = [[random.randint(1, 10) for _ in range(size)] for _ in range(size)]
matrix2 = [[random.randint(1, 10) for _ in range(size)] for _ in range(size)]
# Start timer
start_time = time.time()
# Initialize result matrix with zeros
result = [[0 for _ in range(size)] for _ in range(size)]
# Matrix multiplication using nested loops
for i in range(size):
    for j in range(size):
        for k in range(size):
            result[i][j] += matrix1[i][k] * matrix2[k][j]
# End timer
end_time = time.time()
print("Time taken using Python lists:", end_time - start_time, "seconds")
```

```python
# Create two 1000x1000 matrices
matrix1_np = np.random.randint(1, 10, size=(1000, 1000))
matrix2_np = np.random.randint(1, 10, size=(1000, 1000))

# Start timer
start_time = time.time()

# Matrix multiplication using NumPy
result_np = np.dot(matrix1_np, matrix2_np)
# Or you can use: result_np = matrix1_np @ matrix2_np

# End timer
end_time = time.time()

print("Time taken using NumPy arrays:", end_time - start_time, "seconds")
```
