```python
#1
total_fare = df['Fare'].sum()
print("Total Fare:", total_fare)
```

```python
#3
fare_class_2 = df[df['Pclass'] == 2]['Fare'].sum()
print("Second Class Fare Total:", fare_class_2)
fare_class_3 = df[df['Pclass'] == 3]['Fare'].sum()
print("Third Class Fare Total:", fare_class_3)
```

```python
#4
fare_totals = [fare_class_1, fare_class_2, fare_class_3]
print("Fare Totals by Class:", fare_totals)
```

```python
#5
fare_class_1_prop = fare_class_1 / total_fare
fare_class_2_prop = fare_class_2 / total_fare
fare_class_3_prop = fare_class_3 / total_fare

print(fare_class_1_prop)
print(fare_class_2_prop)
print(fare_class_3_prop)
```

###To answer the question perform following operation:

1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).

2. Calculate the total number of passengers on the Titanic.

3. Count the number of passengers in each age group.

4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.

5. Display the proportion as a percentage.

```python
#1
df['age_group'] = pd.cut(
    df['Age'],
    bins=[0, 18, 64, float('inf')],
    labels=['child', 'adult', 'senior']
)
print(df[['Age', 'age_group']].head(10))
```

```python
#2
total_passengers = len(df)
print("Total Passengers:", total_passengers)
```

```python
#3
age_group_counts = df['age_group'].value_counts()
print("Age Group Counts:")
print(age_group_counts)
```

```python
#4
age_group_proportions = age_group_counts / total_passengers
print("Age Group Proportions:")
print(age_group_proportions)
```

```python
#5
age_group_percentages = age_group_proportions * 100
print("Total passengers:", total_passengers)
print("\nCounts by Age Group:\n", age_group_counts)
print("\nProportions (%):\n", age_group_percentages.round(2))
```
