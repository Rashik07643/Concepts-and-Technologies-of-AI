###Problem 2 - Data Imputations:
##Complete all the following Task:
• Dataset for the Task: "medical_student.csv"
1. Load the provided dataset and import in pandas DataFrame.
2. Check info of the DataFrame and identify column with missing (null) values.
3. For the column with missing values fill the values using various techniques we discussed above. Try
to explain why did you select the particular methods for particular column.
4. Check for any duplicate values present in Dataset and do necessary to manage the duplicate items.
{Hint: dataset.duplicated.sum()}

```python
#1
df = pd.read_csv("/content/drive/MyDrive/Concepts-and-Technologies-of-AI/medical_students_dataset.csv")
print(df.head())
```

```python
#2
print(df.info())
print(df.isnull().sum())
```

```python
#3
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['BMI'] = df['BMI'].fillna(df['BMI'].median())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
print(df.isnull().sum())
```

1.Numeric columns → fill with mean, median, or mode depending on distribution.

2.Categorical columns → fill with mode (most frequent value).

    Mean → average for symmetric numeric data

    Median → middle value for skewed/outlier-heavy numeric data

    Mode → most frequent category for categorical data

By choosing wisely, we minimize bias and keep the dataset realistic after imputation.

```python
#4
print("Number of duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()
print("After removal, duplicate rows:", df.duplicated().sum())
```

###Problem - 1:
Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
median, maximum value, and minimum value of the ’Fare’ column?

```python
df = pd.read_csv("/content/drive/MyDrive/Concepts-and-Technologies-of-AI/Titanic-Dataset.csv")
print(df.head())
```

```python
titanic_subset = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
print(titanic_subset.head())
```

```python
first_class = titanic_subset[titanic_subset['Pclass'] == 1]
print(first_class.head())
```

```python
fare_mean = first_class['Fare'].mean()
fare_median = first_class['Fare'].median()
fare_max = first_class['Fare'].max()
fare_min = first_class['Fare'].min()

print("Fare Statistics for First-Class Passengers:")
print(f"Mean: {fare_mean}")
print(f"Median: {fare_median}")
print(f"Maximum: {fare_max}")
print(f"Minimum: {fare_min}")
```

```python
print(df['Embarked'].unique())
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, embarked_dummies], axis=1)
df = df.drop('Embarked', axis=1)
print(df.head())
```

###Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual-
ization to show how the survival distributions vary by gender.

```python
mean_survival_by_sex = df.groupby('Sex')['Survived'].mean()
print(mean_survival_by_sex)
```

```python
import matplotlib.pyplot as plt
```

```python
mean_survival_by_sex.plot(kind='bar', color=['pink','lightblue'], title='Mean Survival Rate by Gender')
plt.ylabel('Mean Survival Rate')
plt.show()
```
