```python
# Daily time (in hours): [study, entertainment, sleep]
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
```

```python
#Task 1: Classify Study Time:
low_study = []
moderate_study = []
high_study = []
for record in time_data:
    study_hours = record[0]  
    if study_hours < 3:
        low_study.append(study_hours)
    elif 3 <= study_hours <= 5:
        moderate_study.append(study_hours)
    else:
        high_study.append(study_hours)
print("Low Study Hours (< 3):", low_study)
print("Moderate Study Hours (3–5):", moderate_study)
print("High Study Hours (> 5):", high_study)
```

```python
# Task 2: Based on Data – Count study time categories:
low_days = len(low_study)
moderate_days = len(moderate_study)
high_days = len(high_study)
print("Number of days with low study time:", low_days)
print("Number of days with moderate study time:", moderate_days)
print("Number of days with high study time:", high_days)
```

```python
#Task 3: Convert Study Hours to Minutes:
study_minutes = [] 
for record in time_data:
    study_hours = record[0]  
    minutes = study_hours * 60
    study_minutes.append(minutes)
print("Study hours converted to minutes:", study_minutes)
```

```python
# Task 4: Analyze Average Time Use
study_hours = []
entertainment_hours = []
sleep_hours = []
for record in time_data:
    study_hours.append(record[0])
    entertainment_hours.append(record[1])
    sleep_hours.append(record[2])
avg_study = sum(study_hours) / len(study_hours)
avg_entertainment = sum(entertainment_hours) / len(entertainment_hours)
avg_sleep = sum(sleep_hours) / len(sleep_hours)
print(f"Average Study Hours: {avg_study:.2f} hours")
print(f"Average Entertainment Hours: {avg_entertainment:.2f} hours")
print(f"Average Sleep Hours: {avg_sleep:.2f} hours")
```

```python
# Task 5: Visualization - Study vs Sleep Pattern
import matplotlib.pyplot as plt

# Extract study and sleep hours from time_data
study_hours = [record[0] for record in time_data]
sleep_hours = [record[2] for record in time_data]
plt.scatter(study_hours, sleep_hours, color='blue', marker='x')
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Scatter Plot of Study vs Sleep Hours")
plt.grid(True)
plt.show()
```

```python
#Task 1 - Sum of Nested Lists:
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Total sum of nested list:", result)
```

```python
#Task 2 - Generate All Permutations of a String:
def generate_permutations(s):
    """
    Generates all unique permutations of a given string using recursion.

    Args:
        s (str): The input string.

    Returns:
        list: A sorted list containing all unique permutations of the string.
    """
    def permute(chars, start, result):
        if start == len(chars) - 1:
            permutation = ''.join(chars)
            result.add(permutation)  
        else:
            for i in range(start, len(chars)):
                chars[start], chars[i] = chars[i], chars[start]
                permute(chars, start + 1, result)
                chars[start], chars[i] = chars[i], chars[start]
    result_set = set()
    permute(list(s), 0, result_set)
    return sorted(list(result_set))  
print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
#Task 3 - Directory Size Calculation:
# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

def calculate_directory_size(directory):
    """
    Calculates the total size of a directory, including all nested subdirectories.

    Args:
        directory (dict): A dictionary where keys are file names (with sizes as integers)
                          or folder names (with nested dictionaries).

    Returns:
        int: Total size of all files in the directory and its subdirectories in KB.
    """
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            # If value is a dictionary, recursively calculate subdirectory size
            total_size += calculate_directory_size(value)
        else:
            # If value is a number, add it as file size
            total_size += value
    return total_size

# Test the function
total = calculate_directory_size(directory_structure)
print(f"Total directory size: {total} KB")
```

```python
#Task 1 - Coin Change Problem:
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic
    programming.
    This function solves the coin change problem by determining the fewest number of
    coins from a given set of coin denominations that sum up to a target amount. The
    solution uses dynamic programming(tabulation) to iteratively build up the minimum
    number of coins required for each amount.
    Parameters:
    coins (list of int): A list of coin denominations available for making change. Each
    coin denomination is a positive integer.
    amount (int): The target amount for which we need to find the minimum number of coins
    . It must be a non-negative integer.
    Returns:
    int: The minimum number of coins required to make the given amount.
    If it is not possible to make the amount with the given coins, returns -1.
    Example:
    >>> min_coins([1, 2, 5], 11)
    3
    >>> min_coins([2], 3)
    -1
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print("Minimum coins needed:", result)
```

```python
#Task 2 - Longest Common Subsequence (LCS):
def longest_common_subsequence(s1, s2):
    """
    Returns the length of the longest sequence that appears in both strings.

    s1: first string
    s2: second string

    Looks for the longest common subsequence (LCS) and returns its length.
    """
    m, n = len(s1), len(s2)
    # Create DP table with (m+1) x (n+1) initialized to 0
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    # Build the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m][n]
# Test the function
s1 = "abcde"
s2 = "ace"
lcs_length = longest_common_subsequence(s1, s2)
print("Length of LCS:", lcs_length)
```

```python
#Task 3 - 0/1 Knapsack Problem:
def knapsack(weights, values, capacity):
    n = len(weights)
    # Create DP table with (n+1) x (capacity+1) initialized to 0
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Build the table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                # Option 1: take item i, Option 2: skip item i
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                # Cannot take item i, skip it
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

# Test case
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

max_value = knapsack(weights, values, capacity)
print("Maximum value:", max_value)
```
